### seaborn: statistical data visualization

```python
import pandas as pd
import seaborn as sns
sns.color_palette()
```

```python
sns.set_palette('dark')
sns.color_palette()
```

```python
with open('sales_data_sample.csv') as f:
    print(f)
```

```python
sales = pd.read_csv('sales_data_sample.csv', encoding='ISO-8859-1')
```

```python
sales.shape
```

```python
sales.info()
```

```python
sales.describe()
```

```python
sales = sales.dropna(axis=1)
```

```python
sales.columns
```

```python
sales.groupby('DEALSIZE').size()
```

```python
sales.groupby('PRODUCTLINE').size()
```

```python
sales.groupby('COUNTRY').size()
```

```python
# relational plot
sns.relplot(x='SALES', y='MSRP', data=sales)
```

```python
sns.relplot(x='SALES', y='MSRP', data=sales, hue='DEALSIZE')
```

```python
sns.relplot(x='SALES', y='MSRP', data=sales, hue='DEALSIZE', style='STATUS')
```

```python
sns.relplot(x='SALES', y='MSRP', data=sales, size='PRICEEACH')
```

```python
sns.relplot(x='SALES', y='MSRP', data=sales, size='PRICEEACH', sizes=(25,200))
```

```python
# line plot
sns.relplot(x='PRICEEACH', y='SALES', data=sales, kind='line')
```

```python
sns.relplot(x='YEAR_ID', y='SALES', data=sales, kind='line')
```

```python
sns.relplot(x='YEAR_ID', y='SALES', data=sales, kind='line', ci=None) # confidence interval = None
```

```python
sns.relplot(x='YEAR_ID', y='SALES', data=sales, kind='line', ci='sd') # confidence interval = standard deviation
```

```python
sns.relplot(x='YEAR_ID', y='SALES', data=sales, kind='line', estimator=None)
```

```python
sns.relplot(x='YEAR_ID', y='SALES', data=sales, kind='line', estimator=sum)
```

```python
sns.relplot(x='YEAR_ID', y='SALES', data=sales, kind='line', hue='DEALSIZE')
```

```python
sns.relplot(x='YEAR_ID', y='SALES', data=sales, kind='line', hue='DEALSIZE', style='STATUS')
```

```python
sns.relplot(x='YEAR_ID', y='SALES', data=sales, kind='line', hue='DEALSIZE', style='STATUS', ci=None)
```

```python
sns.relplot(x='YEAR_ID', y='SALES', data=sales, kind='line', hue='DEALSIZE', col='STATUS', ci=None, col_wrap=2)
```

```python
# categorical plot
sns.catplot(x='DEALSIZE', y='SALES', data=sales)
```

```python
sns.catplot(x='DEALSIZE', y='SALES', data=sales, jitter=False)
```

```python
sns.catplot(x='DEALSIZE', y='SALES', data=sales.query('PRICEEACH < 65'), jitter=False)
```

```python
sns.catplot(x='DEALSIZE', y='SALES', data=sales.query('PRICEEACH < 65'), kind='swarm')
```

```python
sns.catplot(x='SALES', y='DEALSIZE', data=sales.query('PRICEEACH < 65'), kind='swarm')
```

```python
# facet
sns.relplot(x='SALES', y='MSRP', data=sales, col='DEALSIZE')
```

```python
sns.relplot(x='SALES', y='MSRP', data=sales, col='STATUS', col_wrap=2)
```

```python
sns.relplot(x='SALES', y='MSRP', data=sales, col='STATUS', hue='DEALSIZE', col_wrap=2)
```

```python
sns.relplot(x='SALES', y='MSRP', data=sales, col='DEALSIZE', row='STATUS', hue='PRODUCTLINE')
```

```python
# violin plot
sns.catplot(x='DEALSIZE', y='SALES', data=sales, kind='violin')
```

```python
sns.catplot(x='DEALSIZE', y='SALES', data=sales, hue='STATUS', kind='violin')
```

```python
sns.catplot(x='DEALSIZE', y='SALES', data=sales, hue='STATUS', kind='violin', inner='stick')
```
