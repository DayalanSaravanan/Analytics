### seaborn: statistical data visualization

```python
import pandas as pd
import seaborn as sns
sns.color_palette()
```

```python
sns.set_palette('dark')
sns.color_palette()
```

```python
with open('sales_data_sample.csv') as f:
    print(f)
```

```python
sales = pd.read_csv('sales_data_sample.csv', encoding='ISO-8859-1')
```

```python
sales.shape
```

```python
sales.info()
```

```python
sales.groupby('DEALSIZE').size()
```

```python
sales.groupby('COUNTRY').size()
```

```python
# relational plot
sns.relplot('SALES','MSRP',data=sales)
```

```python
sns.relplot('SALES','MSRP',data=sales,hue='DEALSIZE')
```

```python
sns.relplot('SALES','MSRP',data=sales,hue='DEALSIZE',style='STATUS')
```

```python
sns.relplot('SALES','MSRP',data=sales,size='PRICEEACH')
```
